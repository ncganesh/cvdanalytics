```python
from elasticsearch import Elasticsearch
from elasticsearch_dsl import Search
import pandas as pd
from datetime import datetime,timedelta
```

## ES - Get All Data 

```python

client =  Elasticsearch(hosts=["http://3.238.229.207:9200/"])
s = Search(using=client, index="twitter_india_covid")

df = pd.DataFrame([hit.to_dict() for hit in s.scan()])
```

# Get Data based on timefframe

```python
result = client.search(index="twitter_india_covid", body={
        "query": {
            "range": {
                "@timestamp": {
                    "gte": datetime.utcnow() - timedelta(hours=20),
                    "lt": datetime.utcnow()
                }
            }
        },
        # ensure that we return all docs in our test corpus
        "size":10000
    })

lis = []
for item in result['hits']['hits']:
    lis.append(item['_source'])
```
